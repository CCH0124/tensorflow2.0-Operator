```python
from tensorflow.keras import models, layers
```

- Dense 密集層
    - 每個神經元都會與上一層的每個神經元連接，又稱**全連接層**
    - input_dim
        - 只有第一層才設定
        - 輸入 *feature* 數量
- compile
    - optimizer 指定優化器
        - [官方](https://www.tensorflow.org/api_docs/python/tf/keras/optimizers/Optimizer)
    - loss 指定損失函數
        - 會依照要解決的問題而配置，也要與輸出層的激活函數配合
        - [官方](https://www.tensorflow.org/api_docs/python/tf/keras/losses)
    - metrics 指定評估
        - [官方](https://www.tensorflow.org/api_docs/python/tf/keras/metrics)

|問題類型|輸出層激活函數|損失函數|類型描述|
|---|---|---|---|
|二元分類|sigmoid|binary_crossentropy|資料分為兩類|
|多元分類|softmax|categorical_crossentropy|資料分為多類|
|多標籤分類|sigmoid|binary_crossentropy|將資料貼上 0 到多個標籤|
|迴歸求值|Non|mse|預測一個數值|
|回歸求 0~1 值|sigmoid|mse 或 binary_crossentropy|預測一個數值|

```python
model = models.Sequential()
model.add(layers.Dense(4, activation='relu', input_dim=3))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```
