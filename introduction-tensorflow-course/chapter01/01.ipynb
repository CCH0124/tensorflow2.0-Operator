```python
import tensorflow as tf
```

```python
xs = tf.constant([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=tf.float16)
ys = tf.constant([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=tf.float16)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error') # loss 用來調整訓練和實際的誤差
```

```python
model.fit(xs, ys, epochs=500)
```

```python
print(model.predict([10.0]))
```
